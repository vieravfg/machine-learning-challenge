```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features.drop("koi_disposition", axis=1)
y = selected_features["koi_disposition"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=20)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaled = MinMaxScaler().fit(X_train)
X_train_scaled= X_scaled.transform(X_train)
X_test_scaled = X_scaled.transform(X_test)
```

# Train the Model


```python
from sklearn.ensemble import RandomForestClassifier
random_f = RandomForestClassifier(n_estimators=200)
random_f.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {random_f.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {random_f.score(X_test_scaled, y_test)}")
```

```python
features = X.columns
sorted(zip(random_f.feature_importances_, features), reverse=True)
```

# Hyperparameter Tuning

Use GridSearchCV to tune the model's parameters



```python
#find the parameters currently used
print(f'Parameters in use: {random_f.get_params()}')
```

```python
# Create the GridSearchCV model (Testing 6 different Hyperparameters)
from sklearn.model_selection import GridSearchCV
param_grid = {'bootstrap': [True],
    'max_depth': [125, 150, 175],
    'max_features': [1, 2, 3],
    'n_estimators': [100, 200, 300]}
grid = GridSearchCV(random_f, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Training score:
print(f'Training score:{grid.score(X_train_scaled, y_train)}')
# Testing score:
print(f'Testing score:{grid.score(X_test_scaled, y_test)}')
```
